```python
# Upload file in Google Colab
from google.colab import files
uploaded = files.upload()

import pandas as pd

# 1. Read CSV file
df = pd.read_csv("Movies.csv")
print("Full Dataset:\n", df)

# 2. Display first and last rows
print("\nFirst 3 rows:\n", df.head(3))
print("\nLast 3 rows:\n", df.tail(3))

# 3. Check rows, columns, and data types
print("\nShape (rows, columns):", df.shape)
print("\nColumns:\n", df.columns)
print("\nData Types:\n", df.dtypes)

# 4. Select a single column
print("\nTitle column:\n", df["Title"])

# 5. Select multiple columns
print("\nTitle and Genre columns:\n", df[["Title", "Genre"]])

# 6. Filter data (IMDB_Rating < 9)
filtered = df[df["IMDB_Rating"] < 9]
print("\nMovies with rating < 9:\n", filtered)

# 7. Find missing values
print("\nMissing values (True/False):\n", df.isnull())
print("\nMissing values count:\n", df.isnull().sum())

# 8. Fill missing values with mean
df["IMDB_Rating"] = df["IMDB_Rating"].fillna(df["IMDB_Rating"].mean())
print("\nAfter filling missing values:\n", df)

# 9. Remove duplicate rows
df = df.drop_duplicates()
print("\nAfter removing duplicates:\n", df)

# 10. Simple movie rating analysis
print("\nHighest Rating:", df["IMDB_Rating"].max())
print("Average Rating:", df["IMDB_Rating"].mean())
print("Lowest Rating:", df["IMDB_Rating"].min())

# 11. Sort data by ReleaseYear
df_sorted = df.sort_values("ReleaseYear")
print("\nSorted by ReleaseYear:\n", df_sorted)

# 12. Calculate average rating
print("\nAverage Rating:", df["IMDB_Rating"].mean())

# 13. Group by Genre
print("\nAverage rating by Genre:\n", df.groupby("Genre")["IMDB_Rating"].mean())

# 14. Add a new column (Rating_Score)
df["Rating_Score"] = df["IMDB_Rating"] * 0.10
print("\nAfter adding Rating_Score column:\n", df)

# 15. Save to new CSV
df.to_csv("Newb.csv", index=False)
print("\nData saved to Newb.csv successfully")
```
